```python
#Replace space with underscore
dataFrame.columns = dataFrame.columns.str.lower().str.replace(' ','_')
dataFrame.columns
dataFrameColumns = list(dataFrame.columns)

#Get non-numeric columns
categoricalColumns = list(dataFrame.dtypes[dataFrame.dtypes == 'object'].index)
categoricalColumns.remove('surname')

#Get numerical columns
numericalColumns = list(filter(lambda x: x not in categoricalColumns, dataFrameColumns))
numericalColumns.remove('surname')

print(numericalColumns)


for c in categoricalColumns:
    dataFrame[c] = dataFrame[c].str.lower().str.replace(' ', '_')
```

```python
dataFrame = dataFrame.fillna(0)
```

```python
#use the train test split to have the data 
dataFrameFullTrain, dataFrameTest = train_test_split(dataFrame,test_size=0.2,random_state=1)
dataFrameTrain, dataFrameValidate = train_test_split(dataFrameFullTrain,test_size=0.25,random_state=1)
dataFrameTrain.reset_index(drop=True)
dataFrameTest.reset_index(drop=True)
dataFrameValidate.reset_index(drop=True)
```

```python
#Get the risk factor and global & group difference to know the churn rate
for c in categoricalColumns:
    print(c)
    dataFrameGroup = dataFrameFullTrain.groupby(c).exited.agg(['mean','count'])
    dataFrameGroup['diff'] = dataFrameGroup['mean'] - globalChurn
    dataFrameGroup['risk'] = dataFrameGroup['mean'] / globalChurn
    display(dataFrameGroup)
    print()
    print()
```

```python

from sklearn.metrics import mutual_info_score
```

```python

def calculate_mi(series):
    return mutual_info_score(series, dataFrameFullTrain.exited)

dataFrameMutualInformation = dataFrameFullTrain[categoricalColumns].apply(calculate_mi)
dataFrameMutualInformation = dataFrameMutualInformation.sort_values(ascending=False).to_frame(name='MI')


display(dataFrameMutualInformation.head())
display(dataFrameMutualInformation.tail())
```
