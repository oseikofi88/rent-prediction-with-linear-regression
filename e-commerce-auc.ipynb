````python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
````

````python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
````

````python
df = pd.read_csv('online_retail_customer_churn.csv')

df.columns = df.columns.str.lower().str.replace(' ', '_')

categorical_columns = list(df.dtypes[df.dtypes == 'object'].index)


#replace spaces with underscores and convert to lowercase 
for c in categorical_columns:
    df[c] = df[c].str.lower().str.replace(' ', '_')
    
#convert numbers to numeric values and to 0 where required

df.annual_income = pd.to_numeric(df.annual_income, errors='coerce')
df.annual_income = df.annual_income.fillna(0)

df.total_spend = pd.to_numeric(df.total_spend, errors='coerce')
df.total_spend = df.total_spend.fillna(0)

df.average_transaction_amount = pd.to_numeric(df.average_transaction_amount, errors='coerce')
df.average_transaction_amount = df.average_transaction_amount.fillna(0)
   
df.target_churn = (df.target_churn == 'true' ).astype(int)
    
````

````python
df.dtypes
````

````python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)

df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train.target_churn.values
y_val = df_val.target_churn.values
y_test = df_test.target_churn.values

del df_train['target_churn']
del df_val['target_churn']
del df_test['target_churn']
````

## Data Split Explanation

**Train, Validation, and Test Sets:**

The code splits your dataset into three subsets:

1. **First split (80% train + val, 20% test):**
   - `df_full_train` = 80% of original data (will be split further)
   - `df_test` = 20% of original data (held out, never seen during training)

2. **Second split (75% train, 25% val from df_full_train):**
   - `df_train` = 60% of original data (75% of 80%)
   - `df_val` = 20% of original data (25% of 80%)

**Final breakdown:**
- `df_train`: 60% — Used to train the model
- `df_val`: 20% — Used to tune hyperparameters & check overfitting
- `df_test`: 20% — Used **only once** at the end to evaluate final model (unbiased estimate)

**Why this matters:**
- Training on `df_train` teaches the model patterns
- Validating on `df_val` tells you if it's overfitting (memorizing instead of learning)
- Testing on `df_test` gives a final, honest performance score (because the model never saw this data)

**Visualized:**
```
Original Dataset (100%)
│
├─ df_full_train (80%)
│  ├─ df_train (60%)        ← model learns from this
│  └─ df_val (20%)          ← model is evaluated on this (hyperparameter tuning)
│
└─ df_test (20%)            ← final test (seen only once at the very end)
```

````python
categorical_columns = ['gender','promotion_response','email_opt_in']
numerical_columns = [
    'customer_id',
    'age',
    'annual_income',
    'total_spend',
    'years_as_customer',
    'num_of_purchases',
    'average_transaction_amount',
    'num_of_returns',
    'num_of_support_contacts',
    'satisfaction_score',
    'last_purchase_days_ago',
]
````

````python

dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical_columns + numerical_columns].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)


model = LogisticRegression()
model.fit(X_train, y_train)
````
